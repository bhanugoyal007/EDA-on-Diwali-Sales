# Importing Dataset and performing Data cleaning and wrangling techniques on the dataset 

```python
#Importing necessary libraries
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from matplotlib import style as style
```

```python
#Importing the dataset
df=pd.read_csv('data.csv', encoding = 'unicode_escape')
```

```python
#Displaying the first 20 records.
df.head(20)
```

```python
#Dropping the Unnecessay columns
df=df.drop(columns=['unnamed1','Unnamed: 15'])
```

```python
df=df.drop(columns=['Status'])
```

```python
df.head()
```

```python
#Displaying the total records in a dataset before removing duplicates
df.shape
```

```python
#Dropping the duplicates in a dataset. 8 Duplicate values are found and removed.
df=df.drop_duplicates()
df
```

```python
#Renaming the 'Amount' column to 'Sales'
df=df.rename(columns={'Amount':"Sales"})
```

```python
#Replacing the values in 'Marital_Status' column to 'Married","Unmarried'
df['Marital_Status']=df['Marital_Status'].replace([1,0],["Married","Unmarried"])
```

```python
#Checking for the Nan values in the columns.
df.isna().sum()
```

```python
#Dropping the Nan values as we have found total 12 Nan values in a dataset which is less than the 5% of total records.
df=df.dropna(subset=['Sales'],how='all',axis=0)
df.head(50)
```

```python
df.isnull().sum()
```

```python
#Getting all the information of the columns for the further analysis
df.info()
```

```python
#Getting all the statistical values of the numerical column in the dataset
df.describe(exclude=('object'))
```
